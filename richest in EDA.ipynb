```python
data=data.replace(['\$',""],regex=True)
data['Total Net Worth']=data.replace(['B',""],regex=True)
data['Total Net Worth']=pd.to_numeric(data['Total Net Worth'],errors='coerce')
```

```python
data.head()
```

```python
data=data.replace(['\$',""],regex=True)
data['$ Last Change']=data.replace(['K',""],regex=True)
data['$ Last Change']=data.replace(['M',""],regex=True)
data['$ Last Change']=data.replace(['B',""],regex=True)

data['$ Last Change']=pd.to_numeric(data['$ Last Change'],errors='coerce')
```

```python
data.head()
```

```python
data=data.replace(['\$',""],regex=True)
data['$ YTD Change']=data.replace(['K',""],regex=True)
data['$ YTD Change']=data.replace(['M',""],regex=True)
data['$ YTD Change']=data.replace(['B',""],regex=True)

data['$ YTD Change']=pd.to_numeric(data['$ Last Change'],errors='coerce')
```
